```python
apple2 = apple.reset_index()['Close']
google2 = google.reset_index()['Close']
apple2.shape , google2.shape
```

```python
def dataset_creator(dataset , time_step = 1):
  X = []
  y = []
  for i in range(0 , len(dataset) - time_step):
    a = dataset[i:(i+time_step) , 0]
    X.append(a)
    y.append(dataset[i+time_step , 0])
  return np.array(X) , np.array(y)
```

```python
apple2 , google2 = np.array(apple2) , np.array(google2)
```

```python
apple2 , google2 = apple2.reshape(-1,1) , google2.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
apple2 = sc.fit_transform(np.array(apple2))
google2 = sc.fit_transform(np.array(google2))
```

```python
apple2.shape , google2.shape
```

```python
time_step = 100
X_apple , y_apple = dataset_creator(apple2 , time_step)
X_google , y_google = dataset_creator(google2 , time_step)
```

```python
X_apple = X_apple.reshape(X_apple.shape[0] , X_apple.shape[1] , 1)
X_google = X_google.reshape(X_google.shape[0] , X_google.shape[1] , 1)
```

```python
y_apple.shape
```

```python
preds_apple = apple_predictor.predict(X_apple)
```

```python
preds_google = google_predictor.predict(X_google)
```

```python
preds_apple = sc.inverse_transform(preds_apple)
```

```python
preds_apple.shape
```
